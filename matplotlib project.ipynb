```python
import matplotlib.pyplot as plt
x = [2,3,4,5,6,8]
y = [2,3,4,5,6,8]
plt.plot(x,y, color = "red")
plt.grid(which = "major", color = "blue")
plt.minorticks_on()
plt.grid(which = "minor", color = "lightblue")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7,8]
y = [1,2,3,4,5,6,7,8]
z = [2,4,6,7,8,9,2,4]
plt.plot(x,y)
plt.plot(x,z)
plt.savefig("twolines.jpeg",format = "jpeg")
```

```python
plt.subplot(1,1,1)
plt.bar(x,y)
plt.bar(x,z)
```

```python
plt.scatter(x,y, color = "g")
plt.scatter(x,z)
plt.show(x,y)
```

```python
values = [1,2,3,4,5]
clrs = ["r","b","g","b","c"]
explode = [0,0,0.6,0,0]
lbls = ["india","united states","Russia","china","Europe"]
plt.pie(values,colors = clrs,labels = lbls,explode = explode) # explode thaniya ethu iruko atha thaniya kondu varum like 0.6 Russia
plt.title("student location")
plt.show()
```

```python
plt.plot([1,2,3,4,5],[2,3,4,5,6])
plt.axis([0,6,0,20])
plt.xlabel("No of years")
plt.ylabel("Growth")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0,5,0.2)
plt.plot(t,t,"r--",t,t**2,"b*",t,t**3,"g^") # oru t -- la varum b star star ra varum 
plt.show()
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
f = np.arange(-3,3,0.2)
plt.plot(f,norm.pdf(f)) #normalisation chart varum
plt.show()
```
