### Data Cleaning

```python
# series mean just one column
df["Cabin"]
```

```python
# it will fill in null value rows
df["Cabin"].fillna("bill", inplace = True)
```

```python
# drop unwanted  column
df.drop(["Embarked"], axis = 1, inplace = True)
```

```python
df.drop(["Cabin"], axis = 1, inplace = True)
```

```python
df
```

```python
df["Age"].fillna("bill", inplace = True)
```

```python
# fill age column using mean value
df["Age"].fillna(df["Age"].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
# to find the mean value of the age
df["Age"].mean()
```

```python
df
```

```python
# fix error wrong value to correct value
df["Sex"] = df["Sex"].replace("male","Male")
df["Sex"] = df["Sex"].replace("female","FeMale")
```

### Data Transformation

```python
#  astype it will change data types name
df["Age"] = df["Age"].astype(int)
```

```python
df.dtypes
```

```python
# create new feature new column
df["Familycount"] = df["SibSp"] + df["Parch"]
df
```

```python
#group by functions (mean(),Median(),Max(),Min(),Count(),Sum())
```

```python
# group the pclass column and store it in group_1b
# it is single groupby column
group_1b = df.groupby("Pclass")
```

```python
# groupby multiplecolumn (0 death rate 1 survived rate)
sur_not = df.groupby(["Survived","Sex"])["PassengerId"].count()
sur_not
```

### Data Visualization

```python
# Bar chart used for categorical value and numerical column
# Line chart used for continues value month wise sales, tempature wise raining
# Scatter chart used to check two numrical column relationship
# Histogram chart for distribution 
#syntax of plot 
# Bar Chart sns.barplot(x="column",y="column",data=df)
# Line plot sns.lineplot(x="column",y="column",data=df)
# Histogram chart sns.histplot(df["column"], bins=10, kde=False)
# Scatter plot sns.scatterplot(x="colname",y="colname",data=df)
```

```python
# figsize it is use to decide plot size
plt.figure(figsize=(8,6))
sns.barplot(x ="Sex", y="Survived", data =df)
plt.title("Gender vs Survived Rate")
plt.xlabel("Gender")
plt.ylabel("Survived")
plt.show()
```

```python
sns.barplot(x ="Sex", y="Survived", data =df ,color = "pink", edgecolor = "red")
plt.show()
```

```python
# age Distribution bins means bar count 20 bars in chart kde create line 
plt.figure(figsize=(8,6))
sns.histplot(df["Age"],bins = 20, color="teal", kde= True)
```

```python
sns.scatterplot(x="Age", y="Fare", data=df)
plt.title("Relationship Between Fare and Age")
```

```python
sns.lineplot(x= "PassengerId",y="Fare", data=df, marker = "o",linestyle="-.")
```
